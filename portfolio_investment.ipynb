```python
import pandas as pd
import investment
```

```python
def get(tickers):
    def data(ticker):
        return (investment.Investment(ticker).investment_data)
    datas = map(data,tickers)
    return(pd.concat(datas, keys=tickers, names=['Ticker', 'Date']))
```

```python
tickers=['aapl','t','c']
```

```python
port_data=get(tickers)
```

```python
port_data.tail()
```

```python
portfolio_data = {ticker:investment.Investment(ticker) for ticker in tickers}
```
